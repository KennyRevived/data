# Sets

Sets are an unordered collection of unique elements

```python
# Create a set
my_set = {'football', 'baseball', 'tennis'}
another_set = set([1, 2, 3, 4])
```

Since sets are unordered we cannot use indexes with them

```python
set_x[0]
```

Since sets do not allow duplicates, one common use of sets is to get unique elements

```python
nset = {1, 2, 1, 3}
print(nset)
```

```python
my_list = [30, 2, 2, 14, 30, 1]
my_uniqe_set = set(my_list)
my_uniqe_set
```

Below are some of the commonly used functions avalaible for sets

```python
# len
len(my_set)
```

```python
len(another_set)
```

```python
# in
print('tennis' in my_set)
print('cricket' in my_set)
print(1 in another_set)
```

With sets we can perform many mathematical operations also as below

```python
set_x = {2, 4, 6 ,8}
set_y = {3, 6, 9, 12}
```

```python
# union
combined = set_x.union(set_y) # combine the elements of two sets
print(combined)
```

```python
comb_set = my_set.union(another_set) # set can contain elements from different types
comb_set
```

```python
# intersection
common = set_x.intersection(set_y) # get commom elements between two elements
print(common)
```

```python
comm_set = my_set.intersection(another_set) # no common elements- return empty set
comm_set
```

```python
# difference
diff = set_x.difference(set_y) # get elements from x that are not in y
print(diff)
```

```python
diff_opp = set_y.difference(set_x) # get elements from y that are not in x
diff_opp
```

```python
# update
set_x.update(set_y) # add elements of y to x
print(set_x)
```

```python
my_set.update(another_set)
my_set
```

```python
# isdisjoint
set_x.isdisjoint(set_y) # check if x and y are disjoint set  i.e. they have no element in common
```

```python
my_set.isdisjoint({5,6,7}) # no elements in common-> return True
```

```python
# issubset
set_x.issubset(set_y) # check if x is subset of y
```

```python
my_set.issubset(another_set)
```

```python
# issuperset
set_x.issuperset(set_y) # check if x is superset of y
```

```python
my_set.issuperset(another_set)
```

```python
# pop
x = set_x.pop() # remove last element of set
print(x)
print(set_x)
```

```python
my_set.pop()
my_set
```

```python
# clear
set_x.clear() #  remove all elements of x
print(set_x)
```

```python
my_set.clear()
my_set
```

# Excercise
1. Create a set *flowers* from the list: ['Rose', 'Orchids', 'Lily', 'Sunflower'].
2. Creata another set *more_flowers* with items as 'Poppy', 'Vilot', 'Rose'.
3. Get the unique items from both sets.
4. Get the common items between two sets.
5. Get items unique to more_flowers (not in flowers).
6. Get items unique to each set.
7. Check if the set {'Poppy', 'Vilot', 'Rose'} is a subset or a superset of *more_flowers*.
8. From *flowers* remove the items common between the two sets.
9. Append elements from *more_flowers* to *flowers*.
10. Get the count of distinct elements from list [2, 3, 5,1, 3, 4, 1, 2].

# Solution

```python
# Excecise 1
flowers = set(['Rose', 'Orchids', 'Lily', 'Sunflower'])
flowers
```

```python
# Excecise 2
more_flowers = {'Poppy', 'Vilot', 'Rose'}
```

```python
# Excecise 3
flowers.union(more_flowers)
```

```python
# Excecise 4
flowers.intersection(more_flowers)
```

```python
# Excecise 5
more_flowers.difference(flowers)
```

```python
# Excecise 6- option 1
flowers_unq = flowers.difference(more_flowers)
more_flowers_unq = more_flowers.difference(flowers)
unq_all = flowers_unq.union(more_flowers_unq)
unq_all
```

```python
# Excecise 6- option 2
# use the function symmetric_difference (https://www.w3schools.com/python/ref_set_symmetric_difference.asp)
more_flowers.symmetric_difference(flowers)
```

```python
# Excercise 7
print({'Poppy', 'Vilot', 'Rose'}.issubset(more_flowers))
print({'Poppy', 'Vilot', 'Rose'}.issuperset(more_flowers))
```

```python
# Excercise 8- option 1
flowers = flowers.difference(more_flowers)
flowers
```

```python
# Excercise 8- option 2
# use the function difference_update (https://www.w3schools.com/python/ref_set_difference_update.asp)
flowers.difference_update(more_flowers)
```

```python
# Excercise 9
flowers.update(more_flowers)
flowers
```

```python
# Excercise 10
s = set([2, 3, 5,1, 3, 4, 1, 2])
len(s)
```
