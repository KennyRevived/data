###Fundamental Building Blocks
Variables are used to hold a piece of information. This variable can be reused later on.
can be a number or some text or more complex information

```python
my_program = "Hello World"
print("Hello World")
```

```python
class_name = 'Class 8-1'
math_marks = 80
science_marks = 89
english_marks = 76
```

```python
total_marks = math_marks + science_marks + english_marks
```

```python
print(class_name)
print(total_marks)
```

Let's update the value of the variable

```python
math_marks = 78
total_marks = math_marks + science_marks
print(total_marks)
```

###Data Types
Every variable has a data type, for example: integer, string, boolean, etc.

Here are some examples:

```python
print(type(35))
print(type('Dog'))
print(type('35'))
print(type(45.7))
print(type(True))
```
