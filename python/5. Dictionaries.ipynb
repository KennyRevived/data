# Dictionaries

Dictonaries are a collection of Key:Value pair items.

Let's see how to create a dictionary in Python.

```python
# 2 ways to create a dictionary

my_dict = {'name': 'Becky', 'age': 23, 'job':'marketing'}
my_another_dict = dict([('lang1','English'), ('lang2','French')])
```

Below are the most common used methods for dictionaries:

```python
my_dict['name'] # returs the value of the key passed
```

```python
my_dict['country'] # missing key throws an error
```

```python
len(my_dict) # number of key value pairs
```

```python
my_dict['name'] = 'Claire' # changing a value
my_dict
```

```python
# items

my_dict.items() # get a list of all key-value pairs in dictionary
```

```python
my_another_dict.items()
```

```python
# keys

my_dict.keys() # get a list of all keys in dictionary
```

```python
my_another_dict.keys()
```

```python
# values

my_dict.values() # returns all the values in the list
```

```python
my_another_dict.values()
```

```python
# get

my_dict.get('name') # get the value of a specific key
```

```python
print(my_another_dict.get('lang3')) # missing key- returns none (deafult)

print(my_dict.get('salary',-1)) # defining a value to return if the specified key is missing
```

```python
# update

another_dict = {'city': 'Paris'}
my_dict.update(another_dict) # append items from another dictionary to the dictionary
my_dict
```

```python
my_another_dict.update({'lang3': 'Hebrew'})
my_another_dict
```

```python
# pop

city = my_dict.pop('city') # removes the specified key from dictionary
print(city)
print(my_dict)
```

```python
lang4 = my_another_dict.pop('lang4') # missing key- returns key erorr (deafult)
print(lang4)
print(my_another_dict)
```

```python
lang4 = my_another_dict.pop('lang4',0) # defining a value to return if the specified key is missing
print(lang4)
print(my_another_dict)
```

```python
fruit_dict = {'apple': 3, 'banana': 2, 'cherry': 5}
qty = fruit_dict.pop('apple') # keep the value
print(qty)
print(fruit_dict)
```

```python
# popitem

item = my_dict.popitem()  # removes the last key from dictionary
print(item)
print(my_dict)
```

```python
lang = my_another_dict.popitem()
print(lang)
print(my_another_dict)
```

```python
# clear

my_dict.clear() # removes all items from the dictionary
print(my_dict)
```

```python
my_another_dict.clear()
my_another_dict
```

```python
# in
fruit_dict = {'apple': 3, 'banana': 2, 'cherry': 5} # test if the given key exists in the dictionary
print('banana' in fruit_dict)
print('orange' in fruit_dict)
```

1. Create a dictionary of the below cars and their launch year
* Ford Model T 1908
* Volkawagon Beetle 1938
* Toyota Corolla 1966
* Pointac Firebird 1978
* Chevrolet Panther 1969
2. Print the launch year of 'Ford Model T'.
3. Change the year of 'Pointac Firebird' to 1977.
4. Create another dictionary as {'Dodge Charger' :1968 }. Append this dictionary to the above dicionary.
5. Remove the 'Toyota Corolla' entry from the dictionary.
6. Print all the car names in the dictionary.
7. Print the number of items in dictionary.
8. Check if 'MG Hector' exists in the dictionary.
9. Get the car name with the earliest launch year (Hint: use the optional 'key' parameter of the min function).

# Solution

```python
# Excercise 1
cars = {'Ford Model T' : 1908
, 'Volkawagon Beetle' : 1938
, 'Toyota Corolla' : 1966
, 'Pointac Firebird' : 1977
, 'Chevrolet Panther' : 1969}
```

```python
# Excercise 2
cars['Ford Model T']
```

```python
# Excercise 3
cars['Pointac Firebird'] = 1977
cars
```

```python
# Excercise 4
cars.update({'Dodge Charger' :1968 })
cars
```

```python
# Excercise 5
cars.pop('Toyota Corolla')
print(cars)
```

```python
# Excercise 6
print(cars.keys())
```

```python
# Excercise 7
len(cars)
```

```python
# Excercise 8
'MG Hector' in cars.keys()
```

```python
# Excercise 9
min(cars, key=cars.get)
```

```python
# Excercise 10
cars['Chevrolet Camaro'] = cars.pop('Chevrolet Panther')
cars
```
