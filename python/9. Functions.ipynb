# Functions

Functions are a block of code that can be used to perform a task.<br>
Using functions we can make any piece of code reusable, since we can easily execute the same code block multiple times by just using function name.<br>
Function can accept inputs, known as parameters.<br>
Functions may or may not return values.<br>

Syntax to define a function:
> def <function_name> ( \<param1> = \<defalut value>, \<param2> = \<defalut_value>):
>> #code block<br>
>> \<statement1> <br>
>> \<statement2> <br>
>> return \<expression><br>


```python
def print_me(var1 = 'Nothing!!!'):
    print(var1)

x = 'I am a cat'
print_me(x)
print_me()
```

We can add DocString to a fuction to add some description to the function.

```python
def add(a, b):
    '''
    This function adds two numbers
    '''
    return a + b

add(1 ,2)
```

Functions can return a result also.

```python
def double_me(param1):
    return param1 * 2

x = double_me(10)
print(x)

print(double_me('I am doubled...'))
```

Function can include conditions.

```python
def add(a, b):
    '''
    This function adds two numbers
    '''
    if type(a) == int or type(a) == float:
       if type(b) == int or type(b) == float:
         return a + b

    return ('pass me ints')

print(add(1 ,2))
print(add(10 ,2.5))
print(add(2, 'a'))
```

### Lambda functions

we can create functions without names as well. These anonymous functions are called as Lambda functions.<br>
These functions are created using keyword lamda as below:
>lambda \<param_list>: \<expression>

Lambda funcions can take any number of paramters but can contain only one expression.

We can create a simple add function as


```python
def add(a,b):
    return a + b
```

In lamda form same can be written as

```python
lambda a, b: a+b
```

let's see how we can use lamda expressions

```python
sum = lambda x,y : x+y
sum(2 ,10)
```

```python
is_palindrome = lambda s: s[::-1] == s
print(is_palindrome('rotor'))
print(is_palindrome('rotors'))
```

```python
my_list = [1, 2, 3, 4 ,5]
even = lambda x: x%2 and 'odd' or 'even'
[even(x) for x in my_list]
```

```python
my_list = [x for x in range(30)]
l3 =  list( filter(lambda x: x % 3 == 0, my_list) )
print(l3)
```

```python
# Excercise 1
def check_str(my_str: str):
    if my_str.isalpha():
        print('String is alphabetic only')
    elif my_str.isnumeric():
        print('String is numeric only')
    elif my_str.isalnum():
        print('String is both alphabetic and numeric')

check_str('qwerty')
check_str('1234')
check_str('qwer123')
```

```python
# Excercise 2
def say_hello(n):
    for i in range(n):
        print('Hello')

say_hello(2)
```

```python
# Excercise 3
def get_tax(income):
    if income < 0:
         return 'invalid entry'
    elif 0 < income and income < 10000:
         return 0
    elif 10000 <= income and income < 100000:
        return income * 0.1
    elif 100000 <= income :
        return income * 0.2


print(get_tax(-250))
print(get_tax(250))
print(get_tax(2500000))
```

```python
# Excercise 4
s = 'I am a short string'

def count_char(text):
    v = 0
    c = 0
    for ch in text:
        if ch in ['a', 'e', 'i', 'o', 'u']:
            v = v + 1
        else:
            c = c + 1
    print('there are {} vovwel(s)'.format(v))
    print('there are {} consonant(s)'.format(c))

count_char(s)
```

```python
# Excercise 5
list_a = [1, 2, 3, 4]
list_b = [1, 2, 3, 4, 5]

even_len = lambda x: len(x) % 2 == 0

print(even_len(list_a))
print(even_len(list_b))
```

```python
# Excercise 6
def sum_diff(num1, num2):
    if (type(num1) == int or type(num1) == float) and (type(num2) == int or type(num2) == float):
        return (num1+num2, abs(num1-num2))
    else:
        return 'ERROR! Both inputs should be numbers.'

print(sum_diff(1, 2))
print(sum_diff('1', 2))
```

```python
# Excercise 7
def check_palindrome(text:str):
    text = str(text)
    if(text[:] == text[::-1]):
        return True
    else:
        return False

print(check_palindrome('rotor'))
print(check_palindrome('test'))
print(check_palindrome('racecar'))
```

```python
# Excercise 8
def check_anagram(str1, str2):
    l1 = sorted(str1)
    l2 = sorted(str2)
    if l1 == l2:
        return True
    else:
        return False

print(check_anagram('EARTH', 'HEART'))
print(check_anagram('BAD', 'BAT'))
```

```python
# Excercise 9
def sum_digits(num):
    s = 0
    n = num
    while n > 0:
        s += n%10
        n = n//10
    return s

print(sum_digits(123))
print(sum_digits(555))
```

```python
# Excercise 10
def list_range(l:list):
    l_min = min(l)
    l_max = max(l)
    l_range = l_max - l_min
    return l_min, l_max, l_range
list_range([9, 3, 6, 8])
```

```python
# Excercise 11:

def usd_converter(usd):
    usd_to_eur = round(usd*0.93, 2)
    usd_to_jpy = round(usd*147.41, 2)
    usd_to_mxn = round(usd*17.52, 2)
    return usd_to_eur, usd_to_jpy, usd_to_mxn

usd = float(input('please enter a monetary value (USD): '))
eur, jpy, mxn = usd_converter(usd)
print('EUR: {}, JPY: {}, MXN: {}'.format(eur, jpy, mxn))
```
