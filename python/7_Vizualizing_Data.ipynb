###Matplotlib
The matplotlib library provides a wide range of plotting functions and customization options.<br>

Import matplotlib:

```python
import matplotlib.pyplot as plt
```

Let's see an example:

```python
pen_species = pen.species.value_counts() # count number by species
pen_species
```

```python
plt.show() # show the graph
plt.figure(figsize=(5,3)) # set the figure size (width, height)

pen_species.plot(kind='bar') # plot a bar graph

plt.title('Count By Species') # add a title to the graph
plt.xlabel('species') # add label to the x axis
plt.ylabel('count') # add label to the y axis
```

####**Extra**- multiple plots in one figure (subplots)

In Matplotlib, a Figure is the top-level container that holds everything in your plot. It can contain one or more Axes. <br>
Figure: The entire window or page that displays your plot. It's like the canvas on which you paint your visualizations. <br>
Axes: The individual plotting area or space where your data is actually plotted. <br>
A Figure can have multiple Axes. Axes include things like the x-axis, y-axis, labels, titles, and the actual plotted data.


```python
pen_gender = pen.sex.value_counts() # count number by gender
pen_gender
```

```python
f, axes = plt.subplots(1, 2, figsize = (10,4)) # plot multiples graphs on the same figure

pen_species.plot(kind='bar', ax=axes[0]) # use the axis parameter to specify on which axis to draw the graph
pen_gender.plot(kind='bar', ax=axes[1])
```

###Count plot
The countplot() method creates bar plots that display the count of observations in each category of a categorical variable.

```python
# import plotting libraries
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# below command ensure the graphs are visible in the Notebook
%matplotlib inline
```

```python
# show the counts by species, but this time use the countplot method
plt.figure(figsize=(5,3))
sns.countplot(data=pen, x='species')
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data=pen, y='species') # horizontal graph
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data=pen, x='species', palette='pastel') # we can set the colors using the palette parameter
```

```python
# now, let's use matplotlib to add title, custom axis labels etc.

plt.figure(figsize=(5,3))

sns.countplot(data=pen, x='species', palette='pastel')

plt.title('Count By Species')
plt.xlabel('Species')
plt.ylabel('Count')
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data=pen, x='species', hue='sex') # define subgroups by the hue parameter
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data=pen, x='species', hue='island') # define subgroups by the hue parameter
```

###Bar plot
The barplot() method is used to aggregate categorical data according to certain methods (default- mean). <br>
To use this plot, we choose a categorical column for the x-axis, a numerical column for the y-axis, and an aggregation function to apply to the values in each category (using the "estimator" parameter).

```python
# import plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the penguins dataset
pen = sns.load_dataset("penguins")
pen.head()
```

```python
plt.figure(figsize=(5,3))
sns.barplot(pen, x='island', y='body_mass_g') # average body mass by island
plt.title('average body mass by island')
```

```python
plt.figure(figsize=(5,3))
sns.barplot(pen, x='species', y='bill_length_mm', estimator='median') # median of bill length by species
plt.title('median bill length by species')
```

```python
plt.figure(figsize=(5,3))
sns.barplot(pen, x='species', y='body_mass_g', estimator='sum') # sum of body mass by species
plt.title('sum body mass by species')
```

```python
plt.figure(figsize=(5,3))
sns.barplot(pen, x='island', y='flipper_length_mm', estimator='max') # max flipper length by island
plt.title('max flipper length by island')
```

```python
plt.figure(figsize=(5,3))
sns.barplot(pen, x='island', y='flipper_length_mm', estimator='min') # min flipper length by island
plt.title('min flipper length by island')
```

```python
plt.figure(figsize=(5,3))
sns.barplot(pen, x='island', y='flipper_length_mm', estimator=len) # count
plt.title('count by island')
```

```python
sns.barplot(pen, x='species', y='body_mass_g', hue='sex', palette='flare') # average body mass by species and gender
plt.title('average body mass by species and gender')
```

###Cat plot
The catplot() method is a versatile function for creating different kinds of categorical plots. <br>
The primary advantage of catplot is its ability to easily split the data to multiple subplots based on one or more categorical variables.

```python
# import plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the penguins dataset
pen = sns.load_dataset("penguins")
pen.head()
```

```python
sns.catplot(data=pen, x='species', col='island', kind='count') # kind=count -> countplot
```

```python
sns.catplot(data=pen, x='species', col='island', kind='count', height=4, aspect=0.8) # use the height and aspect parameters to control figure size
```

```python
sns.catplot(data=pen, x='species', col='island', row='sex', kind='count', height=3, aspect=1) # you can specify row and col to divide the data by 2 parameters
```

```python
sns.catplot(data=pen, x='species', y='body_mass_g', col='island', kind='bar', height=4, aspect=0.8) # kind=bar-> barplot
```

###Lineplot
The lineplot() method is a simple way to create line plots using the Seaborn library. <br>
Line plots are useful for showing trends and patterns in data over a continuous scale, like time or another variable. <br>
If there are multiple observations of the y variable at a certain x level, the function will aggregate the y values (default- mean).

```python
# import plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the flights dataset
flights = sns.load_dataset('flights')
flights.head()
```

```python
plt.figure(figsize=(5,3))
sns.lineplot(data=flights, x='year', y='passengers')
plt.title('average number of passengers by year')
```

```python
plt.figure(figsize=(5,3))
sns.lineplot(data=flights, x='year', y='passengers', estimator='sum') # use the estimator parameter to change default aggregation function
plt.title('number of passengers by year')
```

```python
sns.lineplot(data=flights, x='month', y='passengers', hue='year', estimator='sum')  # use the hue parameter to draw multiple lines
plt.title('number of passengers by year and month')
```

```python
# select only flights from 1949
flights_1949 = flights[flights['year']==1949]
flights_1949
```

```python
# no aggregation
plt.figure(figsize=(5,3))
sns.lineplot(data=flights_1949, x='month', y='passengers')
plt.title('number of passengers by month in 1949')
```

###Scatter plot
The scatterplot() method helps you visualize the relationship between two numerical variables by placing points on a graph. <br>
Each point represents a pair of values from the two variables, making it easy to see patterns, trends, or correlations in your data.<br>
Scatter plots are useful for understanding how changes in one variable relate to changes in another.

```python
# import plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the diamonds dataset (and sample 1000 datapoints randomly)
diamonds = sns.load_dataset('diamonds').sample(1000)
diamonds.head()
```

```python
sns.scatterplot(data=diamonds, x='price', y='carat')
plt.title('price vs. carat')
```

```python
# hue- grouping to subsets of points

specific_clr = diamonds[(diamonds['clarity']=='I1') | (diamonds['clarity']=='SI2') | (diamonds['clarity']=='VS1')]

sns.scatterplot(data=specific_clr, x='price', y='carat', hue='clarity')
```

```python
# to fix the legend:
hue_order = ['I1', 'SI2', 'VS1']
sns.scatterplot(data=specific_clr, x='price', y='carat', hue='clarity', hue_order=hue_order)
legend_labels = ['I1', 'SI2', 'VS1']
plt.legend(title='Clarity', labels=legend_labels)
```

```python
# size - the size of the points
sns.scatterplot(data=diamonds, x='price', y='carat', size='y', hue='y')
plt.title('price vs. carat')
```

###FacetGrid
The FacetGrid() method is a handy feature in Seaborn that allows you to create a grid of subplots based on one or more categorical variables.<br>
This makes it easy to visualize and compare different subsets of your data in separate plots within the same figure.

```python
# import plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the flights dataset
flights = sns.load_dataset('flights')
flights.head()
```

```python
g = sns.FacetGrid(flights, col='year') # set the division to seperate plots
g.map(sns.lineplot, 'month', 'passengers') # apply a ploting function to each plot
```

```python
# we can arrange the plots differently by the col_wrap parameter
g = sns.FacetGrid(flights, col='year', col_wrap=4)
g.map(sns.lineplot, 'month', 'passengers')
```

```python
# load the diamonds dataset
diamonds = sns.load_dataset('diamonds').sample(1000)
diamonds.head()
```

```python
g = sns.FacetGrid(diamonds, col='cut', col_wrap=3) # set the division to seperate plots
g.map(sns.scatterplot, 'price', 'carat') # apply a ploting function to each plot
```
