# Basic functions
Python has a rich collection of functions available for lists. Let's take a look to some of them below.

```python
# create a list
my_list = [2, 4, 5, 10]
fruit_list = ['banana', 'orange', 'apple']
```

```python
len(my_list) # length of list is the number of elements in the list
```

```python
len(fruit_list) # length of list is the number of elements in the list
```

```python
# min

print(min(my_list)) # smallest element of the list
print(min(fruit_list)) # list of strings- lowest value, according to alphabetical order
```

```python
# max

print(max(my_list)) # largest element of the list
print(max(fruit_list)) # list of strings- highest value, according to alphabetical order
```

```python
my_list.index(5) # index of the first occurrence of x in s (at or after index i and before index j)
```

```python
# if an element is not found in the specified range we get an error
my_list.index(5, 3, 5)
```

```python
# count

print(my_list.count(2)) # total of times an element appears in the list
print([1,2,3,3,4].count(3))
print(['a',2,3,'3a',4].count('a'))
```

```python
# append- appends an element at the end of list

my_list.append(50) # int
print(my_list)

fruit_list.append('mango') # string
print(fruit_list)
```

```python
my_list.extend([100,200]) # adds the elements of another list to current list
my_list
```

```python
ex_list = ['cherry', 'pineapple'] # adds the elements of another list to current list
fruit_list.extend(ex_list)

fruit_list
```

```python
my_list.insert(3, 500) # inserts a new element at the given index
my_list
```

```python
fruit_list.insert(1, 'grape') # inserts a new element at the given index
fruit_list
```

```python
a = my_list.pop() # removes the last element of the list and returns it
print(my_list)
print(a)
```

```python
a = my_list.pop(1) # removes the element at the given index from the list and returns it
print(my_list)
print(a)
```

```python
my_list.remove(2) # removes the first occurence of the given element
my_list
```

```python
dup_list = [1,1,2,1,4]
dup_list.remove(1) # removes the first occurence of the given element

dup_list
```

```python
my_list.reverse() # reverses the order of the elements in the list
my_list
```

```python
let_list = ['a','b','c','d','e'] # reverses the order of the elements in the list
let_list.reverse()

let_list
```

```python
my_list.sort() # sorts the elements of the list in ascending order
my_list
```

```python
fruit_list.sort() # sorts the elements of the list in ascending alphabetical order
fruit_list
```

```python
my_list.sort(reverse = True) # sorts the elements of the list in descending order
my_list
```

```python
my_list.clear() # removes all the elements from list
print(my_list)
```

### List Comprehension
List Comprehension makes accessing each elements of list very easy.

```python
# raise every element in a matrix to the power of 2
num_list = [1, 2, 3, 4, 5]
squared_num = [x**2 for x in num_list]
print(squared_num)
```

```python
# convert every element in a list to lowercase
str_list = ['Dog','CAT','MouSe']
lower_list = [s.lower() for s in str_list]
print(lower_list)
```

```python
# create a list with the length of each element in another list
len_list = [len(s) for s in str_list]
print(len_list)
```

```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
matrix[0][0]
```

```python
# print the first element of every row in a matrix
first_val = [row[0] for row in matrix]
first_val
```
